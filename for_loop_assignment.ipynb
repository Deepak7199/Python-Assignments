## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
n = int(input("Enter a number: "))

# Assume number is prime
is_prime = True

# Special case: 1 is not prime
if n <= 1:
    is_prime = False
else:
    # Check divisibility from 2 to n-1
    for i in range(2, n):
        if n % i == 0:   # If divisible, not prime
            is_prime = False
            break        # Exit the loop early

# Print result
if is_prime:
    print(n, "is a prime number.")
else:
    print(n, "is not a prime number.")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
# write your code here
N = int(input("Enter a number: "))

print("Prime numbers between 1 and", N, "are:")

# Loop through all numbers from 2 to N
for num in range(2, N + 1):
    flag = True   # Assume current number is prime

    # Check divisibility from 2 to num-1
    for i in range(2, num):
        if num % i == 0:
            flag = False   # Not prime
            break          # Exit inner loop

    # If flag is still True, it's prime
    if flag:
        print(num)
```

## Question 18 (Hard)

**Title:** First Number Divisible by 7 and 11

### Problem Statement:
Take N as input and find the first number between 1 and N that is divisible by both 7 and 11. Stop after finding it.

### Approach:
Use logical AND and stop the loop using `break`.


```python
# write your code here
N = int(input("Enter a number: "))

found = False   # To track if such a number exists

# Loop through numbers from 1 to N
for i in range(1, N + 1):
    if i % 7 == 0 and i % 11 == 0:   # Divisible by both
        print("First number divisible by both 7 and 11 is:", i)
        found = True
        break   # Stop the loop once found

# If no such number exists
if not found:
    print("No number between 1 and", N, "is divisible by both 7 and 11.")
```

## Question 19 (Hard)

**Title:** Number Pattern

### Problem Statement:
Take N as input and print a right-angled number pattern using nested loops.

Example:
1
12
123

### Approach:
Use an outer loop for rows and an inner loop for printing numbers.


```python
# write your code here
N = int(input("Enter a number: "))

# Outer loop for rows
for i in range(1, N + 1):

    # Inner loop for printing numbers
    for j in range(1, i + 1):
        print(j, end="")

    # Move to the next line after each row
    print()
```

## Question 20 (Hard)

**Title:** Perfect Number Check

### Problem Statement:
Take a number as input and check whether it is a perfect number or not.

### Approach:
Find sum of all factors using a loop and compare with the number.


```python
# write your code here
num = int(input("Enter a number: "))

sum_of_factors = 0

# Loop to find all proper divisors
for i in range(1, num):
    if num % i == 0:
        sum_of_factors += i

# Check if perfect number
if sum_of_factors == num:
    print(num, "is a Perfect Number")
else:
    print(num, "is NOT a Perfect Number")
```
