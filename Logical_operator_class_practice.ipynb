## Problem 2: Leap Year

**Question:** Check if a given year is a leap year.

**Approach:** Apply the rules: divisible by 4, but centuries must be divisible by 400.

Rules: Year divisible by 4 is leap, but if divisible by 100 it must also be divisible by 400.

```python
# Enter your answer here
year = int(input("Enter the year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
  print("Leap year")
else:
  print("Non leap year")
```

## Problem 3: Grade Bands

**Question:** Print grade category based on marks: Excellent, Good, Average, Fail.

**Approach:** Build multiple conditions using elif to check ranges.

Take marks (0-100) and print:
- 90–100 → Excellent
- 75–89 → Good
- 50–74 → Average
- else → Fail

```python
# Enter your answer here
marks = int(input("Enter your marks: "))

if marks>=90 and marks<=100:
  print("Excellent")
elif marks>=75 and marks<=89:
  print("Good")
elif marks>=50 and marks<=74:
  print("Average")
else:
  print("Fail")
```

## Problem 7: Delivery Charge System

**Question:** Calculate final delivery charge based on prime membership, amount, and distance.

**Approach:** Start with prime-check, then inside it apply nested if blocks.

Inputs:
- order_amount (int)
- is_prime_member (yes/no)
- distance_km (float)

Rules:
- If prime member → Free delivery
- Else:
  - If amount ≥ 500:
    - If distance ≤ 5 km → Free
    - Else → ₹20 extra
  - Else:
    - If distance ≤ 5 km → ₹30
    - Else → ₹50

```python
# Enter your answer here
amount = float(input("Enter order amount of yours: "))
prime_member = input("Enter you are prime member or not[answer only in yes/no]: ")
distance = float(input("Enter your distance in KM: "))

if prime_member=='yes':
    print("Free Delivery")
else:
    if amount>=500:
        if distance<=5:
            print("Free Delivery")
        else:
            print("Rs.20 will be charged extra")
    else:
        if distance<=5:
            print("Rs.30 will be charged extra")
        else:
            print("Rs.50 will be charged extra")
```

## Problem 8: ATM Withdrawal Simulation

**Question:** Simulate ATM rules: multiples of 100, sufficient balance, manager approval.

**Approach:** Check errors first, then nested conditions for amount > 10000.

Inputs:
- balance
- amount

Logic:
- If amount % 100 != 0 → reject (ATM works with ₹100 multiples)
- If amount > balance → reject
- Else:
  - if amount > 10000 → ask for manager approval
  - else → withdraw success

```python
# Enter your answer here
balance = float(input("Enter total balance in your account: "))
amount = int(input("Enter amount you want to withdraw: "))

if amount%100 != 0:
    print("Reject(ATM works with Rs100 multiples)")
elif amount>balance:
    print("Reject")
else:
    if amount>10000:
        print("Ask for manager approval")
    else:
        print("Withdraw success")
```

## Problem 9: Employee Performance Rating

**Question:** Assign performance rating based on sales, feedback, attendance.

**Approach:** Prioritize attendance rule, then apply nested conditions.

Inputs:
- sales (₹)
- feedback (0-5 scale)
- attendance (%)

Logic:
- If attendance < 70 → 'Poor'
- Else:
  - If sales > 100000 AND feedback > 4.5 → 'Excellent'
  - If sales > 70000 OR feedback > 4 → 'Good'
  - Else → 'Average'

```python
# Enter your answer here
sales = int(input("Enter how much sales you have done this week: "))
feedback = float(input("Enter your feedback[on 0-5 Scale]: "))
attendence = int(input("Enter your attendence: "))

if attendence<70:
    print("Poor")
else:
    if sales>100000 and feedback>4.5:
        print("Excellent")
    elif sales>70000 or feedback>4:
        print("Good")
    else:
        print("Average")
```
