# Q15 - Input: Circle calculations

Ask the user to input the radius of a circle (float). Compute and print:
- Circumference = `2 * 3.14159 * radius`
- Area = `3.14159 * radius ** 2`

```python
# Answer here
# Hint: use input(), then convert to int() or float() as needed.
pie = 3.14159
radius = float(input("Enter the radius to get circumference and area of circle : "))
circumference = 2*pie*radius
area = pie*radius**2
print("The circumference of circle is : ",circumference)
print("The area of circle is : ",area)
```

# Q16 - Input: Convert total seconds to hours, minutes, seconds

Ask the user to input a total number of seconds (integer). Convert it to hours, minutes, and seconds and print as `H:M:S`.

```python
# Answer here
# Hint: use input(), then convert to int() or float() as needed.
second = int(input("Enter time in seconds for getting it in hours and minutes : "))
hours = second/3600
minutes = second/60
print("Entered time in hours is : ",hours,"hours")
print("Entered time in minutes is : ",minutes,"minutes")
print("Entered time in seconds is : ",second,"seconds")
print(f"{hours} : {minutes} : {second}")
```

# Q17 - Input: Simple interest calculation

Ask user for `principal`, `rate` (annual percent), and `time` (years). Compute simple interest as `(principal * rate * time) / 100` and print `Interest: <result>`.

```python
# Answer here
principal = float(input("Enter the principal amount : "))
rate = float(input("Enter the rate on which you are giving principal amount : "))
time = float(input("Enter for how much time you are giving principal amount : "))
simple_interest = (principal*rate*time)/100
print("The interest you will get is : ",simple_interest,"Rs.")
```
