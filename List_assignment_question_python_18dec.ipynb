### Hard 5

**Question:**  
Split a list into even and odd lists.

**Approach:**  
Check each element and append accordingly.


```python
# Write your code here

list_1 = [2,4,6,8,1,3,5,7,9]

list_even =[]
list_odd =[]

for i in range(len(list_1)):
    if list_1[i]%2==0:
        list_even.append(list_1[i])
    else:
        list_odd.append(list_1[i])

print(list_even)
print(list_odd)
```

### Additional Medium 1

**Question:**  
Count how many elements are greater than 10 in a list.

**Approach:**  
Loop and compare each element.


```python
# Write your code here

list_1 = [5,41,20,1,8,9,78,21]

greater_10 = []

for i in range(len(list_1)):
    if list_1[i]>10:
        greater_10.append(list_1[i])
print(len(greater_10))

# Without creating new list.
count = 0
for i in list_1:
    if i>10:
        count += 1
print(count)
```
